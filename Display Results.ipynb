```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
model_names = ['GAM', 'RandomForest', 'NadarayaWatson']
initial_accuracies = [99.13, 99.13, 96.58]
engineered_accuracies = [99.61, 99.40, 98.92]

# Set the position of each bar on the x-axis
x = np.arange(len(model_names))
bar_width = 0.35

# Plot the grouped bars
plt.bar(x, initial_accuracies, width=bar_width, label='Initial Data')
plt.bar(x + bar_width, engineered_accuracies, width=bar_width, label='Engineered Data')

# Add labels, title, and ticks
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy of Models with Initial and Engineered Data')
plt.ylim(top = 130)
plt.yticks([0,20,40,60,80,100])
plt.xticks(x + bar_width/2, model_names)
plt.legend(loc='upper right', bbox_to_anchor=(1, 1))

# Add data labels on top of each bar
for i, v in enumerate(initial_accuracies):
    plt.text(i, v + 1, f'{v:.2f}', ha='center', va='bottom')
for i, v in enumerate(engineered_accuracies):
    plt.text(i + bar_width, v + 1, f'{v:.2f}', ha='center', va='bottom')

# Adjust the layout and display the chart
plt.tight_layout()
plt.show()
```
